```python
data = life_expectancy.copy()
```

```python
data = data.dropna(axis = 0,subset = ['Life_expectancy'])
```

```python
data.info()
```

```python
data.mean()
```

```python
data = data.fillna(data.mean())
```

```python
data.info()
```

```python
data.head()
```

```python
data = data.drop('Country',axis = 1)
data.head()
```

```python
data['Status'] = data['Status'].map({'Developing' : 0, 'Developed' :1})
data.head()
```

```python
column_2 = [ 'Year', 'Status',  'Adult_mortality',
       'Infant_deaths', 'Alcohol', 'Percentage_expenditure', 'HepatitisB',
       'Measles', 'BMI', 'Under_five_deaths', 'Polio', 'Total_expenditure',
       'Diphtheria', 'HIV/AIDS', 'GDP', 'Population', 'Thinness_1-19_years',
       'Thinness_5-9_years', 'Income_composition_of_resources', 'Schooling','Life_expectancy']
```

```python
data = data[column_2]
```

```python
data.head()
```

```python
data_preprocessed = data.copy()
```

```python
data_preprocessed.head()
```

```python
data_preprocessed.info()
```

```python
data_with_target = data_preprocessed.copy()
```

```python
target = data_with_target.iloc[:,-1]
```

```python
target.head()
```

```python
target.shape
```

```python
inputs = data_with_target.iloc[:,:-1]
inputs.head()
```

```python
inputs.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(inputs,target)
```

```python
x_train, x_test, y_train,y_test = train_test_split(inputs, target , train_size = 0.8, random_state = 20)
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape,y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators=10,random_state=0)
```

```python
regressor.fit(x_train, y_train)
```

```python
regressor.score(x_train, y_train)
```

```python
pred = regressor.predict(x_test)
```

```python
pred
```

```python
regressor.score(x_train, y_train)
```

```python
plt.scatter(y_test,pred)
plt.xlabel('Actual value',fontsize=20)
plt.ylabel('Predicted Value',fontsize =20)
plt.show()
```

```python
sns.displot((y_test-pred),bins=30)
```

```python
from sklearn import metrics
```

```python
print('MAE:',metrics.mean_absolute_error(y_test,pred))
print('MSE:',metrics.mean_squared_error(y_test,pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,pred)))
```
